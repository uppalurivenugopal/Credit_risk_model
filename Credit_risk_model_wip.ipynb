#Identify Dependent Variables for the Three Models

For PD Model, we have to calculate the probability of default. For that, we need to know whether the borrower has defaulted or not. This is available in the column "Loan_Status" where we have entries like Fully Paid or Charged Off, Current etc. So, this is our dependent variable

For LGD Model, we need to calculate how much loan was recovered after the borrower has defaulted. This information is available in the "Recoveries" column, So, this will be our dependent variable for the LGD Model

For EAD model, we need to calculate the exposure the momemnt the borrower defaulted. This information is available "total_rec_prncp" (Total Recovered Principal) column. So, this becomes our dependent variable. 

#Caution --> Distinctive Feature of the PD model is that all the independent variables need to be categorical. The reason is that it is simpler to convert into scorecard

#Preprocessing Continous Variables - Emp_length

```python
#Identify unique values
loan_data['emp_length'].unique()
#nan indicates missing values
#we need to convert this to numeric. For this, we need to remove year/ years/ < and other charecters. For that we use str.replace
#sequence of steps will be of major importance as we can't start with str.replace('years') - this will not work on 10+ years or 1 Year
```

```python
loan_data['emp_length_int']=loan_data['emp_length'].str.replace('\+ years','') #this will work on 10+ years
loan_data['emp_length_int']=loan_data['emp_length_int'].str.replace('< 1 year',str(0)) #Note that we are now doing replacements in "Emp_int_int column"
loan_data['emp_length_int']=loan_data['emp_length_int'].str.replace('n/a',str(0)) #we are not converting all missing values to string 0
loan_data['emp_length_int']=loan_data['emp_length_int'].str.replace(' years','') 
loan_data['emp_length_int']=loan_data['emp_length_int'].str.replace(' year','') 
```

```python
#check type of variable in the new column that we created
type(loan_data['emp_length_int'][0])
```

```python
#we need to convert this to numeric. For that we will use pd.to_numeric
loan_data['emp_length_int']=pd.to_numeric(loan_data['emp_length_int'])
```

```python
#check/ confirm
type(loan_data['emp_length_int'][0])
```
